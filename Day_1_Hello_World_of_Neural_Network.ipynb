# Neural Network With 1 Neuron

```python
# In classical programming, we need rules and data to create answers.
# With machine learning, we need data and answers to create rules.
```

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Create Model

```python
# Let's create the most simple neural network model which is a neural network with 1 layer and 1 neuron
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# In here, we have data which is symbolled as x
# And we also have answers that is symbolled as y

# The relationship between x and y is Y = 2(X) - 1
# In traditional programming, we need x (data) and rules to create y

# But let's see it with neural network, can the neural network guess the relationship between X and Y ?

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Compile Model

```python
# Compiling the model, optimizer and loss will be explained in the future.
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# And then, let's train our neural network with 100 iterations
model.fit(xs, ys, epochs=400)
```

```python
# We could predict it using model.predict()
print(model.predict([10.0]))
```

```python
# As we can see, the real answer according to rules is 19 (Y = 2X - 1)
# The neural network is close to guessing it, I guess that's not bad.
# From here we could also know that neural network cannot guess the EXACT number.
# The neural network only guess the number that is close to the real answers.
```
